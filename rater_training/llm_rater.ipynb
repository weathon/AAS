```python
from datasets import load_from_disk
ds = load_from_disk("aas_benchmark_final_llm_selected")
```

```python
import numpy as np
ans_map = {-1: 1, 0:0, 1:2}
llm_ans = np.array([ans_map[x] for x in ds["llm_selected"]])
sec_llm_ans = np.array([ans_map[x] for x in ds["sec_llm_selected"]])
```

```python
idx = 1437
print(llm_ans[idx], sec_llm_ans[idx])
ds["image_distorted"][idx]
```

```python
from sklearn.metrics import accuracy_score, cohen_kappa_score
cohen_kappa_score(sec_llm_ans, llm_ans, weights='quadratic') 
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(sec_llm_ans, llm_ans)
```
